#Create functions that separate the 'Win-Loss' colomn into two integer columns, one that has wins and one that has losses

```python
def stringcleanW(x):
    L = x.split("-")
    return int(L[0])
```

```python
def stringcleanL(x):
    L = x.split("-")
    return int(L[1])
```

#use the functions created above and use the apply function to apply it to every row in the data set

```python
cfb_df['Win'] = cfb_df['Win-Loss'].apply(stringcleanW)
```

```python
cfb_df['Loss'] = cfb_df['Win-Loss'].apply(stringcleanL)
```

#Create a win percentage column by dividing the team win column by the total number of games played(win + loss)

```python
cfb_df['win_per'] = cfb_df['Win']/(cfb_df['Win']+cfb_df['Loss'])
```
